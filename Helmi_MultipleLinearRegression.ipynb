```python
#dataset dari http://buku.dioskurn.com/buku1/ch4
#mempersiapkan data
import pandas as pd
train_df = pd.read_csv('bensin2.csv')
train_df
```

```python
#dataset dari http://buku.dioskurn.com/buku1/ch4
#mempersiapkan data
import pandas as pd
test_df = pd.read_csv('bensin2.csv')
test_df
```

```python
#mengelempokan (Preprocessing Dataset)
import numpy as np

X_train = np.array(train_df[['Liter','Penumpang','Kecepatan','Suhu']])
y_train = np.array(train_df[['Kilometer']]).reshape(-1,1)

print(f'X_train :  \n{X_train}')                  
print(f'y_train : \n{y_train}')


X_test = np.array(test_df[['Liter','Penumpang','Kecepatan','Suhu']])
y_test = np.array(test_df[['Kilometer']])

print(f'X_test :  \n{X_test}')                  
print(f'y_test : \n{y_test}')
```

```python
#Melatih model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
                       
print(f'scoring model: {r2_score(y_test,y_pred)}')
```

```python
#mengeluarkan nilai slope hasil perhitungan
model.coef_
```

```python
#mengeluarkan nilai intercept hasil perhitungan
model.intercept_
```

```python
#Prediksi dengan input Liter=30, Penumpang=2, Kecepatan=50, Suhu=10
jarak = model.predict([[30, 2, 50, 10]])
print(f'Jarak = {jarak} Kilometer')
```
